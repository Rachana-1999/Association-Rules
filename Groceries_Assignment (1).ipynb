```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
groc = pd.read_excel("groceries.xlsx")
groc = groc.replace(np.nan, '', regex=True)
groc 
groc
```

```python
groc
transactions = []
for i in range(0, 9834):
    transactions.append([str(groc.values[i,u]) for u in range(1, 31)]) 
```

```python
groceries_series  = pd.DataFrame(pd.Series(transactions))
groceries_series 
groceries_series.columns = ["business"]
groceries_series
```

```python
X = groceries_series['business'].str.join(sep=',').str.get_dummies(sep=',')
X
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(X, min_support=0.005, max_len=3,use_colnames = True)
frequent_itemsets 
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
frequent_itemsets.sort_values
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(20)
rules.sort_values('lift',ascending = False).head(5)
```
